# Importing Dependencies

```python
#%matplotlib inline
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style("whitegrid")
```

# Exploratory Data Analysis 

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# Reading the data
df = pd.read_csv("../Data/bestsellers with categories.csv")
# Getting the top five rows of the data
df.head()
```

```python
# Checking the data types of each features
df.info()
```

```python
# Checking if there is NaN Values
df.isna().sum()
```

```python
# Summary Statistics of the data
df.describe()
```

```python
df["User Rating"].hist(bins=10, color="orange");
```

```python
df["Reviews"].hist(bins=10, color="green");
```

```python
df_log_reviews = np.log10(df["Reviews"])
plt.hist(df_log_reviews, bins=10, color="maroon");
```

## Books with Highest Rating

```python
df[df["User Rating"] == df["User Rating"].max()]
```

## Books with Lowest Ratings  

```python
df[df["User Rating"] < 4]
```

## Book that has most reviews

```python
max_review = df[df["Reviews"] == df["Reviews"].max()]
max_review
```

## Books that has lowest reviews

```python
min_review = df[df["Reviews"] == df["Reviews"].min()]
min_review
```

## Common Authors

```python
common_authors = df.groupby("Author").count().reset_index().sort_values("Name", ascending=False).head(28)
common_authors
```

```python
fig = plt.figure(figsize=(15, 10), dpi = 75)
plt.bar(common_authors["Author"], common_authors["Name"], color=["salmon"])
plt.title("Common Authors")
plt.ylabel("Count")
plt.xlabel("Authors")
plt.xticks(rotation=50);
plt.yticks(rotation=90);
```

## Number of Fiction and Non-Fiction Books 

```python
genre_counts = df["Genre"].value_counts()
genre_counts
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
df["Genre"].value_counts().sort_values(ascending=False).plot(kind="bar",ax=ax, color=["purple", "aquamarine"]);
ax.tick_params(labelsize=14)
ax.set_xlabel("Genre", fontsize=14)
ax.set_ylabel("Counts", fontsize=14);
```

## Correlation of User Rating to Reviews(Non-Transform and Transform)

```python
# Non-Transform reviews
user_rating = df["User Rating"]
reviews = df["Reviews"]
user_rating.corr(reviews)
```

```python
# Transform Reviews
user_rating.corr(df_log_reviews)
```

## Most expensive Books 

```python
expensive_books = df[df["Price"] == df["Price"].max()]
expensive_books
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(df.Year, df.Price, color="crimson")
ax.tick_params(labelsize=14)
ax.set_xlabel("Year", fontsize=14)
ax.set_ylabel("Price", fontsize=14);
```

## User rating based on genre

```python
genre_rating = pd.crosstab(df.Genre, df["User Rating"])
genre_rating
```

```python
sns.boxplot(x="Genre", y=df["User Rating"], data=df);
```

## Duplicated Samples

```python
df_duplicated = df[df.duplicated(["Name"])]
df_duplicated
```

## User Rating and Reviews

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(x=df["User Rating"], y=df["Reviews"], c="salmon");
ax.tick_params(labelsize=14)
ax.set_xlabel("User Rating", fontsize=14)
ax.set_ylabel("Reviews", fontsize=14);
```

## User Rating and Price

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(user_rating, df["Price"], color="lightgreen")
ax.tick_params(labelsize=14)
ax.set_title("User Rating vs Price",fontsize=14)
ax.set_xlabel("User Ratings", fontsize=14)
ax.set_ylabel("Price", fontsize=14);
```

## Reviews and Price

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(df_log_reviews, df.Price, color="salmon")
ax.tick_params(labelsize=14)
ax.set_xlabel("Log of Reviews", fontsize=14)
ax.set_ylabel("Price", fontsize=14);
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(df.Reviews, df.Price, color="lightgreen")
ax.tick_params(labelsize=14)
ax.set_xlabel("Reviews", fontsize=14)
ax.set_ylabel("Price", fontsize=14);
```

## Year and Reviews

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(df.Year, df.Reviews, color="cyan")
ax.tick_params(labelsize=14)
ax.set_xlabel("Year", fontsize=14)
ax.set_ylabel("Reviews", fontsize=14);
```
