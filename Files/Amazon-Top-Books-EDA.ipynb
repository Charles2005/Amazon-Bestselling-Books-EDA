```python
%matplotlib inline
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

# Exploratory Data Analysis 

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# Reading the data
df = pd.read_csv("../Data/bestsellers with categories.csv")
# Getting the top five rows of the data
df.head()
```

```python
# Checking the data types of each features
df.info()
```

```python
# Checking if there is NaN Values
df.isna().sum()
```

```python
# Summary Statistics of the data
df.describe()
```

## Q1: What books have the highest consistent rating? 

```python
df[df["User Rating"] == df["User Rating"].max()]
```

## Q2: What book/s have the lowest rating?  

```python
df[df["User Rating"] < 4]
```

## Q3: Common Authors

```python
df["Author"].value_counts()
```

## Most Reviewed Book/s

```python
df[df["Reviews"] == df["Reviews"].max()]
```

## What books have the lowest reviews

```python
df[df["Reviews"] == df["Reviews"].min()]
```

## Number of Fiction and Non-Fiction Books 

```python
df["Genre"].value_counts()
```

## Is User rating correlates to reviews 

```python
user_rating = df["User Rating"]
reviews = df["Reviews"]
user_rating.corr(reviews)
```

## Most expensive Books 

```python
df[df["Price"] == df["Price"].max()]
```

## User rating based on genre

```python
pd.crosstab(df.Genre, df["User Rating"])
```

## Duplicated Samples

```python
df_duplicated = df[df.duplicated(["Name"])]
df_duplicated
```
